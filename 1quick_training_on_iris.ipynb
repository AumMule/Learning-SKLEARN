```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris_train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1] #[rows ,columns][all,all-last]
```

```python
y=data.iloc[:,-1]
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[6.4	,2.8	,5.6	,2.1]])
```

```python
predictions
```

```python
datatest = pd.read_excel("iris_test.xlsx")
```

```python
x_test = datatest.iloc[:,:-1]
```

```python
prediction = model.predict(x_test.values)
```

```python
prediction
```

```python
prediction_df = pd.DataFrame(prediction, columns=['Predicted_Species'])
```
